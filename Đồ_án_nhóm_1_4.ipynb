4. Phân tích mối liên hệ giữa các biến

     4.1 Biểu đồ (scatter plot)

Biểu đồ phân tán (scatter plot) là một trong những công cụ hữu hiệu và phổ biến nhất trong phân tích dữ liệu, đặc biệt trong việc hiểu và minh họa mối quan hệ giữa hai biến số định lượng. Mục đích chính của biểu đồ phân tán là xác định xem liệu có một mối quan hệ tuyến tính, phi tuyến tính, hoặc không có mối quan hệ nào giữa hai biến. Khi sử dụng scatter plot, chúng ta có thể dễ dàng quan sát mối liên hệ và xác định các mô hình, xu hướng, và các điểm bất thường trong dữ liệu.
Biểu đồ phân tán (scatter plot) là một công cụ phân tích mạnh mẽ trong lĩnh vực chứng khoán, giúp nhà đầu tư hiểu rõ mối quan hệ giữa các biến số định lượng. Trong chứng khoán, các biến số này có thể bao gồm giá cổ phiếu, khối lượng giao dịch, tỷ suất sinh lợi, độ biến động, và nhiều yếu tố khác ảnh hưởng đến hiệu suất của cổ phiếu hoặc danh mục đầu tư. Scatter plot không chỉ giúp phát hiện các xu hướng trong dữ liệu mà còn xác định mối quan hệ tương quan giữa các biến và những điểm bất thường, từ đó đưa ra quyết định đầu tư tốt hơn.
Sử dụng scatter plot để phân tích mối quan hệ giữa các biến: giá mở cửa (open), giá cao nhất (high), giá thấp nhất (low), giá đóng cửa (close), và khối lượng giao dịch (volume).
Kết quả nhận được:

•	Mối quan hệ dương mạnh: Giữa giá mở cửa và giá đóng cửa, cũng như giữa giá cao nhất và giá thấp nhất.

•	Mối quan hệ không rõ ràng: Giữa giá mở cửa, giá đóng cửa với khối lượng giao dịch. Một số ngày có thể có khối lượng giao dịch cao nhưng giá không biến động nhiều, và ngược lại.

•	Phát hiện ngoại lệ (Outliers): Có thể tồn tại những phiên giao dịch mà giá mở cửa hoặc giá đóng cửa có sự chênh lệch lớn với các biến khác, có thể gây ra sự chú ý về các sự kiện bất thường hoặc tin tức.

Kết Luận:

Việc sử dụng biểu đồ phân tán để phân tích mối quan hệ giữa các biến giá cổ phiếu và khối lượng giao dịch sẽ giúp nhà đầu tư có cái nhìn trực quan về cách mà các yếu tố này tương tác với nhau. Kết quả phân tích sẽ hỗ trợ việc ra quyết định đầu tư, cho thấy sự liên kết giữa biến động giá và khối lượng giao dịch, từ đó tạo ra các chiến lược giao dịch hiệu quả hơn.

4.2. Hệ số tương quan (Correlation Matrix)

Ma trận tương quan cung cấp hệ số tương quan giữa tất cả các cặp biến trong tập dữ liệu.
Trong phân tích dữ liệu tài chính, ma trận tương quan (correlation matrix) đóng vai trò quan trọng trong việc hiểu mối quan hệ giữa các biến số. Đặc biệt, khi phân tích giá cổ phiếu, việc nắm bắt tương quan giữa các giá trị như giá mở cửa (open), giá cao nhất (high), giá thấp nhất (low), giá đóng cửa (close), và khối lượng giao dịch (volume) có thể cung cấp thông tin quan trọng về xu hướng thị trường và sự biến động của giá cổ phiếu.

4.3 Hệ số tương quan (correlation coefficient)

Là một chỉ số thống kê dùng để đo lường mức độ và hướng mối quan hệ giữa hai biến số. Giá trị của hệ số này nằm trong khoảng từ -1 đến 1:
•	Hệ số tương quan bằng 1 cho thấy hai biến có mối tương quan thuận hoàn hảo. Khi một biến tăng thì biến kia cũng tăng.
•	Hệ số tương quan bằng -1 cho thấy hai biến có mối tương quan nghịch hoàn hảo. Khi một biến tăng thì biến kia giảm.
•	Hệ số tương quan bằng 0 cho thấy không có mối tương quan tuyến tính giữa hai biến

Độ đồng biến (covariance)

Là một chỉ số thống kê dùng để đo lường mối quan hệ giữa hai biến số. Nó cho biết mức độ mà hai biến số biến động cùng nhau:
-  Đồng biến dương:  cho thấy khi giá cổ phiếu của một ngân hàng tăng, giá cổ phiếu của ngân hàng còn lại cũng có xu hướng tăng.
-  Đồng biến âm:  giá cổ phiếu của một ngân hàng tăng thì giá cổ phiếu của ngân hàng kia giảm.
-  Đồng biến bằng 0: không có mối liên hệ rõ ràng giữa các cổ phiếu.

- Phân Tích Kết Quả
Phân tích kết quả sau khi chạy thuật toán để từ đó hiểu rõ các nhóm cổ phiếu đã phân, đánh giá các nhóm được tạo ra, tìm hiểu các đặc điểm chung của từng nhóm.

o	Ví dụ: Xem xét giá trị trung bình của các chỉ số cho mỗi nhóm để hiểu rõ hơn về hành vi cổ phiếu.

- Trực Quan Hóa Kết Quả
Trực quan hóa kết quả phân tích để dễ dàng hiểu và chia sẻ.
- Rút Ra Kết Luận và Đề Xuất: Tóm tắt các phát hiện và đề xuất hành động dựa trên phân tích.

•	Ví dụ:

o	Nhóm cổ phiếu có sự biến động tương tự có thể phản ứng giống nhau đối với các yếu tố bên ngoài.

o	Các cổ phiếu có tiềm năng tăng trưởng hoặc giảm rủi ro có thể được xác định dựa trên phân tích nhóm.

Kết Quả Nhận Được từ Phân Tích Nhóm trong Phân Tích Giá Cổ Phiếu:

Sau khi thực hiện phân tích nhóm (Cluster Analysis) sử dụng các biến như giá mở cửa, giá cao nhất, giá thấp nhất, giá đóng cửa, và khối lượng giao dịch, bạn có thể rút ra một số kết quả quan trọng như sau:
1.	Phân Nhóm Cổ Phiếu:

Các cổ phiếu được phân nhóm thành những nhóm tương tự dựa trên hành vi giá và khối lượng giao dịch. Mỗi nhóm có thể đại diện cho các cổ phiếu có tính chất và xu hướng tương tự, chẳng hạn như nhóm cổ phiếu có giá tăng mạnh, nhóm cổ phiếu có biến động thấp, hoặc nhóm cổ phiếu có khối lượng giao dịch cao.
2.	Hiểu Biến Động Giá:

Kết quả phân tích cho phép nhận diện những cổ phiếu có sự biến động giá tương tự. Ví dụ, nếu nhóm cổ phiếu của của CTG, BID và AGB trong một khoảng thời gian cụ thể nằm trong cùng một khoảng giá và khối lượng giao dịch tương tự, điều này cho thấy chúng có thể chịu ảnh hưởng từ các yếu tố bên ngoài giống nhau, chẳng hạn như tin tức thị trường hoặc chính sách kinh tế.
3.	Xác Định Xu Hướng:
Kết quả có thể giúp xác định các xu hướng tiềm ẩn trong thị trường.
4.	Ra Quyết Định Đầu Tư:
Việc phân nhóm giúp tìm ra các cổ phiếu có tính chất tương tự, từ đó giúp đưa ra quyết định đầu tư thông minh hơn.
5.	Phát Hiện Các Cổ Phiếu Bất Thường:
Kết quả cũng có thể chỉ ra những cổ phiếu có hành vi bất thường so với phần còn lại của nhóm. Điều này có thể giúp phát hiện những cơ hội đầu tư hoặc cảnh báo về các rủi ro tiềm ẩn.
6.	So Sánh Giữa Các Nhóm:

So sánh hiệu suất giữa các nhóm cổ phiếu khác nhau. Ví dụ, Giả sử trong vòng một năm qua, giá cổ phiếu của VietinBank đã tăng 20%, trong khi BIDV chỉ tăng 10% và Agribank tăng 15%. Điều này cho thấy VietinBank có xu hướng tăng trưởng mạnh mẽ hơn so với hai ngân hàng còn lại.
